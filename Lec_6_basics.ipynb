Exercise 1: List and Loop Practice
Write a Python program that does the following:
1.	Ask the user to input 5 numbers (use a loop to gather inputs).
2.	Store these numbers in a list.
3.	Write a loop that calculates the sum and average of the numbers in the list.
4.	Print the list, the sum, and the average. **bold text**


```python
numbers = []

for i in range(5):
    num = int(input(f"Enter a number{i+1}: "))  # Convert input to an integer
    numbers.append(num)

total_sum= sum(numbers)
average=total_sum/len(numbers)

print(f"Your list of numbers: {numbers}")
print(f"sum of the number:{total_sum}")
print(f"average of the number: {average}")
```

Exercise 4: While Loop and Input Validation
Write a Python program that asks the user to guess a secret number between 1 and 10. The program should:
1.	Continue to ask the user to guess the number until they get it right (use a while loop).
2.	If the guess is wrong, print "Try again!".
3.	If the guess is correct, print "Congratulations, you guessed it!" and exit the loop. **bold text**


```python
import random as Rand
```

```python
x=Rand.randrange(1,10)
while True:
  y=int(input('Guess the number: '))
  if x==y:
    print(f'Congratulations ypou guessed it right the number is {x} and your guess is {y}')
  else:
    print(f'try again! the secret number is {x} and your guess number {y}')
```

Exercise 5: Basic Function Writing
Write a Python function called calculate_total_price that:
1.	Takes two inputs from the user: the price of an item and the quantity.
2.	Calculates the total price.
3.	If the total price is more than 100, apply a 10% discount.
4.	Print the final price after applying any discount. **bold text**


```python
price=int(input('Enter the price: '))
quantity=int(input('Enter the quantity: '))
def function (price,quantity):
  total_price=price*quantity
  if total_price>100:
    total_price=total_price*0.9
    print(total_price)
  return total_price
```
